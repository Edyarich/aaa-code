# Теоретические основы Matplotlib
![Компоненты MatPlotLib](Images/matplotlip_components.jpeg)

- Figure - это контейнер самого верхнего уровня, та область на которой все нарисовано. Таких областей может быть несколько, каждая из которых может содержать несколько контейнеров Axes.
- Axes(Subplot) - это та область на которой чаще всего и отражаются графики (данные в виде графиков), а так же все вспомогательные атрибуты (линии сетки, метки, указатели и т.д.). 
- Каждая область Axes содержит XAxis и YAxis. Они содержат, деления, метки и прочие вспомогательные атрибуты. 

### Подготовка к работе

```python
# Импортируем matplotlib
import matplotlib.pyplot as plt

# Numpy поможет с генерацией некоторых графиков
import numpy as np
```

### Разберём элементы библиотеки на практике
Ниже мы задействуем наиболее полезные настройки для figure и axes.  
Попробуйте поменять значения, чтобы увидеть как изменится отображаемый график.

```python
# Создаём объкт фигуры и axes на его основе
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)

# Раскрашиваем элементы рабочей области
fig.set(facecolor='springgreen')
ax.set(facecolor='lightskyblue')

# Настраиваем наши оси разными способами
ax.set_title('Играем с параметрами!')
ax.set_xlabel('Ось абцисс (XAxis)')
ax.set_xlim([-10, 15])
ax.set(ylabel='Ось ординат (YAxis)', ylim=[0, 2])

# Добавим данные
ax.scatter([-8, -4, 2, 6, 9], [1.5, 0.5, 0.25, 1.75, 1])

plt.show()
```

### Множество графиков на одной фигуре
Иногда нам может понадобиться разместить сразу несколько графиков на одной фигуре. Возможно мы хотим соотнести их друг с другом и для этого нам нужно расставить их определённым образом. А может быть особое расположение требуется только ради эстетики. В любом случае нам понадобится разобраться с subplots чуть получше.

Для каждого сабплота на сколько частей нам нужно поделить фигуру и в какой по счёту ячейке ему разместиться. Нумерация ячеек начинается с 1 и увеличивается слева-напрво, сверху-вниз.  
**fig.add_subplot(2 - Строки, 2 - Столбцы, 1 - Номер ячейки)**

```python
# Создаём фигуру как обычно
fig = plt.figure()

# Создаём и расставляем сабплоты на фигуре
ax_1 = fig.add_subplot(2, 2, 1)
ax_2 = fig.add_subplot(2, 2, 2)
ax_3 = fig.add_subplot(2, 2, 3)
ax_4 = fig.add_subplot(2, 2, 4)

# Для каждого сабплота задаём имя и убираем отметки на осях
ax_1.set(title = 'ax_1', xticks=[], yticks=[])
ax_2.set(title = 'ax_2', xticks=[], yticks=[])
ax_3.set(title = 'ax_3', xticks=[], yticks=[])
ax_4.set(title = 'ax_4', xticks=[], yticks=[])

# Добавим данные для примера работы с сабплотами
ax_4.scatter([-8, -4, 2, 6, 9], [1.5, 0.5, 0.25, 1.75, 1])

plt.show()
```

Для создания однотипных сабплотов можно использовать массивы

```python
# Указываем размерность сетки
fig, axes = plt.subplots(nrows = 2, ncols = 2)

# Обращаться к сабплотам можно по индексам
axes[0, 0].set(title='axes[0, 0]')
axes[0, 1].set(title='axes[0, 1]')
axes[1, 0].set(title='axes[1, 0]')
axes[1, 1].set(title='axes[1, 1]')

# Однотипные действия можно выполнять в цикле
for ax in axes.flat:
    ax.set(xticks=[], yticks=[])

plt.show()
```

**plt.subplots()** без параметров создаёт один сабплот на всю ширину фигуры. По сути это наиболее удобный способ создать график и получить доступ к axes и figure для настройки.  
В большинстве случаев это наиболее предпочтительный способ создания графиков.

```python
# Самый быстрый и универсальный способ сделать график
fig, ax = plt.subplots()
plt.show()
```

### Закрашивание областей под графиками

Иногда нам моджет понадобиться закрасить область, ограниченную одним или несколькими графиками. Например, выделить значения, попадающие в определенный персентиль или подсветить места, где Y одного графика больше Y другого.

В данном примере я не буду услоджнять задачу рассчетами персентиля, об этом мы поговорим позже. Представим что нас интересуют значения, для которых X > 3.

```python
x = np.linspace(-4, 4)
y = -np.power(x, 2) + 16

plt.plot(x, y)
plt.fill_between(x, y, where=(x > 3))

plt.show()
```

Чуть подробнее рассмотрим как работает фильтр. Никакой магии тут нет, выражение x > 3 просто создаёт маску из True и False для каждого значения x:

```python
print(x > 3)
```

В более сложных случаях мы можем захотеть увидеть области, образованные пересечениями нескольких графиков. Для этого мы может указать ещё один параметр.

```python
x = np.linspace(-4, 6, num=100)
y1 = -np.power(x, 2) + 16
y2 = -np.power(x - 2, 2) + 16
y3 = np.minimum(y1, y2)

fig, ax = plt.subplots()
ax.set_ylim([0, 18])
ax.plot(x, y1, label='y1')
ax.plot(x, y2, label='y2')
ax.fill_between(x, y1, y2, color='lightsteelblue')
ax.fill_between(x, y3, 0, color='coral')

plt.legend()
plt.show()
```
