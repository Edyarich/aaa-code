## Продуктовые гипотезы

`Есть гипотеза, что эти изменения приведут к росту конверсии из посетителя в покупателя.`  
1. Conversion Rate = $\dfrac{\sum\text{# contacts}}{\sum\text{# item_views}}$ хотя бы в одной тестовой группе больше, чем в контроле    

`Изменения на поиске будут приводить к изменению структуры трафика на карточке и соответственно к изменению эффекта от улучшений на карточке`  
1. Сравнить конверсию, среднее число поисков, просмотров и контактов в **Контроле** = [Нет изменений, Планировка] и в **Тесте** = [Быстрый поиск/Чекбокс, Планировка]  
2. Тоже самое, но только теперь **Контроль** = [Нет изменений, Нет изменений], **Тест** = [Быстрый поиск/Чекбокс, Нет изменений]  

Зафиксируем ALPHA = 0.01

## EDA

Смотрим на статистики

```python
grouped_data = df.groupby(['changes_in_item', 'changes_in_search'])
grouped_data[['searches', 'item_views', 'contacts']].describe().T
```

Бакетируем

```python
K_GROUPS = 50
df['bucket'] = df['cookie_id'].apply(lambda x: get_bucket(x, K_GROUPS))
```

Средний размер бакета в группе

```python
df.groupby(['changes_in_item', 'changes_in_search'])['bucket'].count() / K_GROUPS
```

Среднее число просмотров в группе

Постройте график распределений, по цвету раздели изменения на поиске, по строкам изменения на карточке.

```python
fig, axes = plt.subplots(3, 2, figsize=(20, 20), sharey=True, sharex=False)

for i, metric in enumerate(df.columns[1:4]):
    for j, item_change in enumerate(df['changes_in_item'].unique()):
        df_slice = df[df['changes_in_item'] == item_change]

        sns.histplot(data=df_slice, x=metric, hue='changes_in_search', ax=axes[i, j], bins=20)

        axes[i, j].set_title(f'Metric: {metric}. Item mode: {item_change}')
```

Визуально изменения между группами не видны из-за тяжёлых хвостов распределений  

Усредним метрики по бакетам

```python
grouped = df.groupby(['changes_in_item', 'changes_in_search', 'bucket'])
grouped_stats = grouped[['searches', 'item_views', 'contacts']].mean().reset_index()

conversion_rate = grouped['contacts'].sum() / grouped['item_views'].sum()
conversion_rate = conversion_rate.reset_index()
conversion_rate.columns = list(conversion_rate.columns[:3]) + ['ctr']
```

Отрисуем средние число поисков, просмотров и контактов ещё раз

```python
fig, axes = plt.subplots(3, 2, figsize=(20, 20), sharey=True, sharex=False)

for i, metric in enumerate(grouped_stats.columns[-3:]):
    for j, item_change in enumerate(grouped_stats['changes_in_item'].unique()):
        df_slice = grouped_stats[grouped_stats['changes_in_item'] == item_change]

        sns.histplot(data=df_slice, x=metric, hue='changes_in_search', ax=axes[i, j], kde=True)

        axes[i, j].set_title(f'Metric: {metric}. Item mode: {item_change}')
```

```python
fig, axes = plt.subplots(3, 3, figsize=(20, 20), sharey=True, sharex=False)

for i, metric in enumerate(grouped_stats.columns[-3:]):
    for j, search_change in enumerate(grouped_stats['changes_in_search'].unique()):
        df_slice = grouped_stats[grouped_stats['changes_in_search'] == search_change]

        sns.histplot(data=df_slice, x=metric, hue='changes_in_item', ax=axes[i, j], kde=True)

        axes[i, j].set_xlabel(None)
        axes[i, j].set_title(f'Metric: {metric}.\nSearch mode: {search_change}')
```

**Вывод:**  
1. Карточка `Add_plan`: + контакты кроме разреза `no_changes`, = просмотры, + поиски кроме разреза `no_changes`, 
2. Поиск `Checkbox`: + контакты, + просмотры, + поиски в разрезе `add_plan`  
   Поиск `Fast_search`: + контакты, + просмотры, + поиски в разрезе `add_plan`

Конверсия по группам

```python
plt.figure(figsize=(16, 9))
ax = plt.gca()

for i, search_change in enumerate(conversion_rate['changes_in_search'].unique()):
    for j, item_change in enumerate(conversion_rate['changes_in_item'].unique()):
        df_slice = conversion_rate[conversion_rate['changes_in_item'] == item_change]
        df_slice = df_slice[df_slice['changes_in_search'] == search_change]
        
        sns.histplot(data=df_slice, x='ctr', ax=ax, kde=True, label=f'{search_change} + {item_change}')

plt.title('CTR by groups')
plt.legend()
plt.show()
```

**Вывод:**  
1. Простое добавление планировки ухудшает конверсию, но если это событие дополнительно сопровождается изменением поиска, то тогда конвресия улучшается  
2. Изменения поиска улучшают конверсию вне зависимости от наличия планировки  

**Общий вывод**  
Изменения поиска становятся более выразительными с добавлением планировки

## Двухфакторная ANOVA

Посмотрим на дисперсию в выборках

```python
grouped_stats.groupby(['changes_in_item', 'changes_in_search']).var()
```

```python
conversion_rate.groupby(['changes_in_item', 'changes_in_search']).var()
```

Видим,что сильных различий нет

Из графиков предыдущего пункта следует, что большинство выборок имеют колонормальное распределение  
С учётом одинакового размера выборок (= 50), можем спокойно применять ANOVA

**Для конверсии**

```python
aov_ctr = pg.anova(
    dv='ctr',
    between=['changes_in_search', 'changes_in_item'],
    data=conversion_rate,
    detailed=True
)
aov_ctr
```

Видим, что фактор поиска по отдельности значим, а фактор добавления планировки -- нет (хоть и на границе)

**Для просмотров**

```python
aov_views = pg.anova(
    dv='item_views',
    between=['changes_in_search', 'changes_in_item'],
    data=grouped_stats,
    detailed=True
)
aov_views
```

**Для поисков**

```python
aov_searches = pg.anova(
    dv='searches',
    between=['changes_in_search', 'changes_in_item'],
    data=grouped_stats,
    detailed=True
)
aov_searches
```

**Для контактов**

```python
aov_contacts = pg.anova(
    dv='contacts',
    between=['changes_in_search', 'changes_in_item'],
    data=grouped_stats,
    detailed=True
)
aov_contacts
```

**Вывод:**  
Значимы лишь изменения в поиске для конверсии, кол-ва просмотров и контактов.  
Число поисков не меняется стат. значимо.

## Критерий Тьюки

```python
ctr_without_plan = conversion_rate[conversion_rate.changes_in_item == 'no_changes']
ctr_with_plan = conversion_rate[conversion_rate.changes_in_item == 'add_plan']
grouped_without_plan = grouped_stats[grouped_stats.changes_in_item == 'no_changes']
grouped_with_plan = grouped_stats[grouped_stats.changes_in_item == 'add_plan']
```

**Конверсия**

Без планировки в карточке

```python
tukey_ctr = pairwise_tukeyhsd(endog=ctr_without_plan['ctr'], groups=ctr_without_plan['changes_in_search'], alpha=0.01)
tukey_ctr.summary()
```

С планировкой в карточке

```python
tukey_ctr_plus = pairwise_tukeyhsd(endog=ctr_with_plan['ctr'], groups=ctr_with_plan['changes_in_search'], alpha=0.01)
tukey_ctr_plus.summary()
```

**Просмотры**

Без планировки в карточке

```python
tukey_views = pairwise_tukeyhsd(endog=grouped_without_plan['item_views'], groups=grouped_without_plan['changes_in_search'], alpha=0.01)
tukey_views.summary()
```

C планировкой в карточке

```python
tukey_views_plus = pairwise_tukeyhsd(endog=grouped_with_plan['item_views'], groups=grouped_with_plan['changes_in_search'], alpha=0.01)
tukey_views_plus.summary()
```

**Контакты**

Без планировки в карточке

```python
tukey_contacts = pairwise_tukeyhsd(endog=grouped_without_plan['contacts'], groups=grouped_without_plan['changes_in_search'], alpha=0.01)
tukey_contacts.summary()
```

С планировкой в карточке

```python
tukey_contacts_plus = pairwise_tukeyhsd(endog=grouped_with_plan['contacts'], groups=grouped_with_plan['changes_in_search'], alpha=0.01)
tukey_contacts_plus.summary()
```

**Точные значения ДИ**  
<u> Напоминание:</u> средний размер бакета в группе составляет 11050 человек

```python
BUCKET_SIZE = 11050
print('Conversion plus')
print(-tukey_ctr_plus.confint[1:3, 1] * 100)
print('Item Views')
print(-tukey_views.confint[1, 1] / BUCKET_SIZE * 100, '%')
print('Contacts')
print(-tukey_contacts.confint[1, 1] / BUCKET_SIZE * 100, '%')
print('Contacts plus')
print(-tukey_contacts_plus.confint[1:3, 1] / BUCKET_SIZE * 100, '%')
```
