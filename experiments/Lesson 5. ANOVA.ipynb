### –†–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –§–∏—à–µ—Ä–∞

–í —Å–ª—É—á–∞–µ –¥–∏—Å–ø–µ—Ä—Å–∏–æ–Ω–Ω–æ–≥–æ –∞–Ω–∞–ª–∏–∑–∞ –¥–ª—è –ø—Ä–æ–≤–µ—Ä–∫–∏ —Ç–µ–æ—Ä–∏–π –∏—Å–ø–æ–ª—å–∑—É–µ—Ç—Å—è —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –§–∏—à–µ—Ä–∞ (F-—Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ).

–ü—É—Å—Ç—å $\chi_1$ –∏ $\chi_1$ ‚Äî –Ω–µ–∑–∞–≤–∏—Å–∏–º—ã–µ —Å–ª—É—á–∞–π–Ω—ã–µ –≤–µ–ª–∏—á–∏–Ω—ã, –∏–º–µ—é—â–∏–µ $\chi^2$-—Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ —Å $\nu_1$ –∏ $\nu_2$ —Å—Ç–µ–ø–µ–Ω—è–º–∏ —Å–≤–æ–±–æ–¥—ã —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤–µ–Ω–Ω–æ, —Ç–æ–≥–¥–∞ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ —Å–ª—É—á–∞–π–Ω–æ–π –≤–µ–ª–∏—á–∏–Ω—ã:

$$F_{\nu_1,\nu_2}=\frac{\displaystyle\frac{\chi_1^2}{\nu_1}}{\displaystyle\frac{\chi_2^2}{\nu_2}}$$
–±—É–¥–µ—Ç –Ω–∞–∑—ã–≤–∞—Ç—å—Å—è —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ–º –§–∏—à–µ—Ä–∞.

–ï—Å–ª–∏ –º—ã –≤–æ–∑—å–º–µ–º –Ω–µ—Å–∫–æ–ª—å–∫–æ —Å—ç–º–ø–ª–æ–≤ –¥–∞–Ω–Ω—ã—Ö –∏–∑ –∏–∑–≤–µ—Å—Ç–Ω–æ–≥–æ –Ω–∞–º –Ω–æ—Ä–º–∞–ª—å–Ω–æ–≥–æ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è, —Ç–æ –æ—Ç–Ω–æ—à–µ–Ω–∏–µ —Å—É–º–º—ã –¥–∏—Å–ø–µ—Ä—Å–∏–π –º–∞–ª–µ–Ω—å–∫–∏—Ö —Å—ç–º–ø–ª–æ–≤ –∫ –æ–±—â–µ–π –¥–∏—Å–ø–µ—Ä—Å–∏–∏ –±—É–¥–µ—Ç —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–æ –∫–∞–∫ $\chi_{n-1}^2$

$$\frac{\sum {(y_n-\overline y)}^2}{\sigma^2}=\chi_{n-1}^2\rightarrow F_{\nu_1,\nu_2}=\frac{\frac{S_1}{\sigma_1}}{\frac{S_2}{\sigma_2}}$$


üëâ –í—Å–µ —ç—Ç–æ –ø–æ–º–æ–≥–∞–µ—Ç –ø–æ–Ω—è—Ç—å, —è–≤–ª—è—é—Ç—Å—è –ª–∏ –≤—ã–±–æ—Ä–∫–∏ –ø–æ–¥—Å—ç–º–ø–ª–∞–º–∏ –∏–∑ –æ–¥–Ω–æ–π –≥–µ–Ω–µ—Ä–∞–ª—å–Ω–æ–π —Å–æ–≤–æ–∫—É–ø–Ω–æ—Å—Ç–∏.  –ï—Å–ª–∏ —è–≤–ª—è—é—Ç—Å—è, —Ç–æ —Ä–∞–∑–ª–∏—á–∏—è –≤ –¥–∞–Ω–Ω—ã—Ö –Ω–æ—Å—è—Ç —Å–ª—É—á–∞–π–Ω—ã–π —Ö–∞—Ä–∞–∫—Ç–µ—Ä.

## –î–≤—É—Ö—Ñ–∞–∫—Ç–æ—Ä–Ω–∞—è ANOVA

–¢–∞–∫–æ–π –¥–∏—Å–ø–µ—Ä—Å–∏–æ–Ω–Ω—ã–π –∞–Ω–∞–ª–∏–∑ —Å—á–∏—Ç–∞–µ—Ç—Å—è –±–ª–∏–∑–∫–æ –∫ —Ç–æ–º—É, –∫–∞–∫ –∏ –æ–¥–Ω–æ—Ñ–∞–∫—Ç–æ—Ä–Ω—ã–π. 


> üìä **–ü—Ä–∏–º–µ—Ä** 
>–û–¥–Ω–æ–≤—Ä–µ–º–µ–Ω–Ω–æ –∏–∑–º–µ–Ω—è–µ–º —à–∏—Ä–∏–Ω—É –ø–æ–∏—Åƒ∏–æ–≤–æ–π –≤—ã–¥–∞—á–∏ –∏ –ø–æ–∏—Åƒ∏–æ–≤—ã–µ —Ñ–∏–ª—å—Ç—Ä—ã (–∫–∞–∫ –Ω–∞ –∏–∑–æ–±—Ä–∞–∂–µ–Ω–∏–∏ –Ω–∏–∂–µ üëá) –∏ —Ö–æ—Ç–∏–º –ø–æ–º–µ—Ä–∏—Ç—å —ç—Ñ—Ñ–µ–∫—Ç —ç—Ç–∏—Ö –∏–∑–º–µ–Ω–µ–Ω–∏–π.<br/>



–ó–¥–µ—Å—å –≤–∏–¥–∏–º –¥–≤–∞ —Ñ–∞–∫—Ç–æ—Ä–∞, –∫–æ—Ç–æ—Ä—ã–µ –æ–¥–Ω–æ–≤—Ä–µ–º–µ–Ω–Ω–æ –≤–ª–∏—è—é—Ç –Ω–∞ —ç–∫—Å–ø–µ—Ä–∏–º–µ–Ω—Ç.

#### 1. –ü–æ—Å—Ç–∞–Ω–æ–≤–∫–∞ –Ω—É–ª–µ–≤–æ–π –≥–∏–ø–æ—Ç–µ–∑—ã

–§–æ—Ä–º—É–ª–∏—Ä—É–µ–º —Ç—Ä–∏ –Ω—É–ª–µ–≤—ã–µ –≥–∏–ø–æ—Ç–µ–∑—ã $H_0$:

$H_0$: –ø–æ–ª–æ–∂–µ–Ω–∏–µ —Ñ–∏–ª—å—Ç—Ä–∞ –Ω–µ –≤–ª–∏—è–µ—Ç –Ω–∞ ƒ∏–æ–Ω–≤–µ—Ä—Å–∏—é –ø–æ–∏—Åƒ∏–∞

$H_0$: —à–∏—Ä–∏–Ω–∞ –ø–æ–∏—Åƒ∏–æ–≤–æ–π ƒ∏–æ–ª–æ–Ωƒ∏–∏ –Ω–µ –≤–ª–∏—è–µ—Ç –Ω–∞ ƒ∏–æ–Ω–≤–µ—Ä—Å–∏—é –ø–æ–∏—Åƒ∏–∞

$H_0$: –Ω–µ—Ç –≤–∑–∞–∏–º–æ–¥–µ–π—Å—Ç–≤–∏—è –º–µ–∂–¥—É —Ñ–∞ƒ∏—Ç–æ—Ä–∞–º–∏


#### 2. –ü—Ä–æ–≤–æ–¥–∏–º –æ—Ü–µ–Ω–∫—É –¥–∏—Å–ø–µ—Ä—Å–∏–∏ –¥–ª—è –∫–∞–∂–¥–æ–≥–æ —Ñ–∞–∫—Ç–æ—Ä–∞

$$MS_A=\frac{SS_A}{df_A}=\frac{bn{\displaystyle\sum_{i=1}^a}{(\overline{X_I}-\overline X)}^2}{a-1}$$

$$MS_B=\frac{SS_B}{df_B}=\frac{an{\displaystyle\sum_{j=1}^b}{(\overline{X_J}-\overline X)}^2}{b-1}$$

#### 3. –í—ã—á–∏—Å–ª—è–µ–º –æ–±—â—É—é –æ—à–∏–±–∫—É –≤ –≥—Ä—É–ø–ø–∞—Ö

$$MS_{error}=\frac{SS_{error}}{df_{error}}=\frac{\sum_{i=1}^a\sum_{j=1}^b{\displaystyle\sum_{l=1}^n}{(X_{ijl}-\overline{X_{IJ}})}^2}{ab(n-1)}$$

#### 4. –û—Ü–µ–Ω–∏–≤–∞–µ–º –≤–∑–∞–∏–º–æ–¥–µ–π—Å—Ç–≤–∏–µ –º–µ–∂–¥—É —Ñ–∞–∫—Ç–æ—Ä–∞–º–∏

$$MS_{AB}=\frac{SS_{AB}}{df_{AB}}=\frac{SS_{cells}-SS_A-SS_B}{df_{cells}-df_A-df_B}$$

–≥–¥–µ $SS_{cells}=\sum_{i=1}^a\sum_{j=1}^bn{(\overline{X_{IJ}}-\overline X)}^2,\;df_{cells}=ab-1$

#### 5. –î–µ–ª–∞–µ–º –≤—ã–≤–æ–¥ –æ –Ω–∞–ª–∏—á–∏–∏ —Å—Ç–∞—Ç–∏—Å—Ç–∏—á–µ—Å–∫–∏ –∑–Ω–∞—á–∏–º—ã—Ö –æ—Ç–ª–∏—á–∏–π –º–µ–∂–¥—É –≥—Ä—É–ø–ø–∞–º–∏
