Инструментальная переменная - это третья переменная Z, используемая в регрессионном анализе, когда у вас есть эндогенные переменные - переменные, на которые влияют другие переменные в модели. ## Классическая линейная регрессия для случая с одной переменной

- В прошлый раз мы с вами вспомнили основные формулы, распределения и операции с матрицами. Все это нам будет необходимо, чтобы построить модель и убедиться, что она удовлетворяет некому набору свойств
- Собственно, одним из краеугольных камней эконометрики является линейная модель регрессии и метод наименьших квадратов (МНК)

- Давайте предположим, что у нас есть некий набор точек $Y$ и $X$. Пока что будем считать, что у нас только одна переменная $X$.
- Мы предполагаем, что их взаимосвязь описывается линейной функцией: $Y = \alpha + \beta \cdot X + \varepsilon$, где $\varepsilon$ - это некая случайная компонента (случайный член). Про нее поговорим чуть позже
- Нам хочется понять взаимосвязь между этими двумя переменными. Например, как изменение опыта работы влияет на заработную плату, как изменение цены на продукт влияет на объем продаж и т.п. Конечно, чтобы ответить на подобные вопросы недостаточно одной переменной и простой линейной модели, но пока что разберем такую спецификацию.

- Итак, у нас есть эти два набора точек. Мы можем с вами отобразить их на плоскости

- Поэтому мы хотим подобрать такую функцию $Y = f(X)$, которая "наилучшим" способом описывает зависимость $Y$ от $X$
- В рамках предположения, что у нас только одна переменная и **линейная** зависимость, функция выглядит как: $f(X) = \hat{\alpha} + \hat{\beta} \cdot X$
- Соответственно, наша цель подобрать "наилучшие" параметры $\hat{\alpha} \text{ и } \hat{\beta}$

### МНК

- Есть две основных причины, почему используется именно этот метод (эта функция потерь) для линейной регресии:
    - легко минимизировать, так как это просто квадратичная функция 
    - хорошие статистические свойства, простота математических выводов делают возможным построить развитую теорию, позволяющую провести тщательную проверку различных статистических гипотез
- Из минусов стоит выделить, что этот метод чувствителен к выбросам

- Итак, в чем же собственно состоит идея МНК
- Как мы с вами видели ранее, у нас есть некий набор пар точек $(X_i, Y_i)$ 
- Учитывая наши предположения, нам нужно провести через это "облако" пар точек некую линию
- Вопрос сводится к тому, какую именно линию

- В рамках МНК нам нужно минимизовать квадрат отклонений расчитанной величины и фактической величины (другими словами, минимизировать квадрат ошибки/остатка, $e$), т.е.:
$$ L = \sum_{i=1}^{n} e_{i}^2 \rightarrow \min_{\hat{\alpha}, \hat{\beta}} \\
\text{ или }\\
L = \sum_{i=1}^{n}(Y_i - f(X_i))^{2} \rightarrow \min_{\hat{\alpha}, \hat{\beta}} \\
\text{ или } \\ 
L = \sum_{i=1}^{n}(Y_i - \hat{\alpha} - \hat{\beta}X_i)^{2} \rightarrow \min_{\hat{\alpha}, \hat{\beta}}$$
- Также, вы могли еще встречать такую запись: $RSS = \sum_{i=1}^{n} e_{i}^2$, где $RSS$ - residual sum of squares, сумма квадратов ошибок/остатков
- Надеюсь, что вы помните, как это сделать с помощью производных.

- Мой совет - попробуйте сами руками вывести дома, но, конечно, мы сделаем это сейчас вместе

$$\frac{\partial L}{\partial \hat{\alpha}} = -2\sum_{i=1}^{n} (Y_i - \hat{\alpha} - \hat{\beta}X_i) = 0 \\
\frac{\partial L}{\partial \hat{\beta}} = -2\sum_{i=1}^{n} X_i(Y_i - \hat{\alpha} - \hat{\beta}X_i) = 0$$

- Константа $(-2)$ не влияет на равенство, поэтому можем убрать ее. Получим следующую систему:
$$\sum_{i=1}^{n} (Y_i - \hat{\alpha} - \hat{\beta}X_i) = 0 \\
\sum_{i=1}^{n} X_i(Y_i - \hat{\alpha} - \hat{\beta}X_i) = 0$$

- Выразим $\hat{\alpha}$ из первого уравнения:
$$\sum_{i=1}^{n} (Y_i - \hat{\alpha} - \hat{\beta}X_i) = 0 \\
\sum_{i=1}^{n} Y_i - \sum_{i=1}^{n} \hat{\alpha} - \sum_{i=1}^{n} \hat{\beta}X_i = 0 \\
\sum_{i=1}^{n} Y_i - n \cdot \hat{\alpha} - \hat{\beta}\sum_{i=1}^{n} X_i = 0 \\
\hat{\alpha} = \frac{1}{n} \sum_{i=1}^{n} Y_i - \frac{1}{n} \hat{\beta}\sum_{i=1}^{n} X_i \\
\hat{\alpha} = \bar{Y} - \hat{\beta} \bar{X}$$

- Подставим полученную оценку константы во второе равенство:
$$\sum_{i=1}^{n} X_i(Y_i - \bar{Y} + \hat{\beta} \bar{X} - \hat{\beta}X_i) = 0 \\
\sum_{i=1}^{n} X_i Y_i - \bar{Y}\sum_{i=1}^{n} X_i + \hat{\beta} \bar{X} \sum_{i=1}^{n} X_i - \hat{\beta}\sum_{i=1}^{n} X_i^2 = 0 \\
\hat{\beta} = \frac{\sum_{i=1}^{n} X_i Y_i - \bar{Y}\sum_{i=1}^{n} X_i}{\sum_{i=1}^{n} X_i^2 - \bar{X} \sum_{i=1}^{n} X_i}$$

- **Дома**
- Докажите, что 
$$\frac{\sum_{i=1}^{n} X_i Y_i - \bar{Y}\sum_{i=1}^{n} X_i}{\sum_{i=1}^{n} X_i^2 - \bar{X} \sum_{i=1}^{n} X_i} = \frac{\sum_{i=1}^{n} (X_i - \bar{X})(Y_i - \bar{Y})}{\sum_{i=1}^{n} (X_i - \bar{X})^2}$$

- В конечном итоге получим, что:
$$ \hat{\alpha} = \bar{Y} - \bar{X}\hat{\beta} \\
\hat{\beta} = \frac{\sum_{i=1}^{n}{x_i y_i}}{\sum_{i=1}^{n}{x_i^2}}$$
где $x_i = X_i - \bar{X} \text{ и } y_i = Y_i - \bar{Y}$

- Не буду выводить формулы дисперсии этих оценок
- Кому будет интересно, они есть в учебниках
$$var(\hat{\alpha}) = \sigma^2 \frac{\sum_{i=1}^{n} X^2_i}{n \sum_{i=1}^{n} x_i^2} \\
var(\hat{\beta}) = \frac{\sigma^2}{\sum_{i=1}^{n} x_i^2} $$
где $\sigma^2$ - дисперсия случайного члена
- На самом деле мы не знаем настоящую дисперсию случайного члена, поэтому пользуются следующей оценкой:
$$\hat{\sigma}^2 = \frac{RSS}{n-k}$$
где k - количество независимых переменных, регрессоров (в нашем случае $k = 1$)

## Теорема Гаусса-Маркова

- Фух, вот теперь мы умеем "руками" считать коэффициенты парной регрессии
- Однако, вопрос в том, являются ли вообще оценки, полученные с помощью МНК, "хорошими"

- Ну, конечно, нам опять надо сначала определить, что мы считаем "хорошей" оценкой
- В курсе по статистике у вас уже затрагивалась эта тема, но давайте вспомним

- Есть три основные метрики "качества" оценки:
    - Состоятельность
    - Несмещенность
    - Эффективность

- Как ни странно, но начнем мы не с состоятельности :)

### Несмещенность
- Формально: 
    - Допустим, у нас есть оценка коэффициента бета: $\hat{\beta}$. Эта оценка будет считаться несмещенной, если $E(\hat{\beta}) = \beta$
- Неформально:
    - Мы хотим, чтобы в среднем наша оценка попадала в настоящее значение
- Давайте пример ниже

```python
good_estimator = []
bad_estimator = []
for i in range(1000):
    # Есть какой то датасет
    a = np.random.normal(loc = 0, scale = 1, size = 1000)
    # Хотим оценить математическое ожидание
    # Вы же помните, что собой представляет несещенная оценка математического ожидания, да?
    good_estimator.append(np.mean(a))
    bad_estimator.append(np.max(a) - np.min(a))
```

```python
plt.figure(figsize=(15, 10))
plt.hist(good_estimator, bins='auto')
plt.axvline(x=0, color = 'red')
plt.show()
```

```python
plt.figure(figsize=(15, 10))
plt.hist(bad_estimator, bins='auto')
plt.axvline(x=0, color = 'red')
plt.show()
```

```python
# А че по регрессии?
good_beta = []
bad_beta = []
for i in range(10000):
    X = np.random.normal(loc = 0, scale = 1, size = 1000)
    eps = np.random.normal(loc = 0, scale = 2, size = 1000)
    Y = 10 + 2 * X + eps
    good_beta.append(np.cov(X, Y)[0][1]/np.var(X))
    bad_beta.append(np.cov(X, Y)[0][1]/np.var(X) + np.cov(X, eps+0.5*X)[0][1]/np.var(X))
```

```python
plt.figure(figsize=(15, 10))
plt.hist(good_beta, bins='auto')
plt.axvline(x=2, color = 'red')
plt.show()
```

```python
plt.figure(figsize=(15, 10))
plt.hist(bad_beta, bins='auto')
plt.axvline(x=2, color = 'red')
plt.show()
```

### Эффективность
- Допустим, у нас есть две несмещенные оценки. Как нам с вами понять, какая из них лучше.
- Опять же, давай рассмотрим пример ниже

```python
beta1 = []
beta2 = []
for i in range(1000):
    X = np.random.normal(loc = 0, scale = 1, size = 1000)
    eps = np.random.normal(loc = 0, scale = 2, size = 1000)
    Y = 10 + 2 * X + eps
    beta1.append(np.cov(X, Y)[0][1]/np.var(X))
    beta2.append(np.mean(np.random.normal(loc = 2, scale = np.var(eps)/np.var(X)+1, size = 1000)))
```

```python
bins = np.linspace(min(np.min(beta1), np.min(beta2)), max(np.max(beta1), np.max(beta2)), 100)
plt.figure(figsize=(15, 10))
plt.hist(beta1, bins=bins, label = 'beta1')
plt.hist(beta2, bins=bins, label = 'beta2')
plt.legend(loc = 'best')
plt.axvline(x=2, color = 'black')
plt.show()
```

```python
np.mean(beta2)
```

```python
np.var(beta1)
```

```python
np.var(beta2)
```

- Другими словами, эффективной считается оценка, которая обладает наименьшей дисперсией

### Состоятельность
- В практических примерах у нас с вами нет возможности проводить много-много экспериментов (обычно)
- Однако мы с вами в ходе одного эксперимента получаем всё новые данные
- Тогда встает вопрос, а наши оценки могут "исправиться", если мы будем продолжать увеличивать количество данных
- Пожалуй, наиболее показательный пример, это оценка дисперсии (которая, смещенная): $\frac{1}{n}\sum_{i=1}^{N}(X_i - \bar{X})^2$

```python
sigma_biased = []
sigma_unbiased = []
for i in range(1000):
    X = np.random.normal(loc = 0, scale = 1, size = 10)
    sigma_biased.append(1/len(X)*np.sum((X-np.mean(X))**2))
    sigma_unbiased.append(1/(len(X)-1)*np.sum((X-np.mean(X))**2))
```

```python
bins = np.linspace(min(np.min(sigma_biased), np.min(sigma_unbiased)), max(np.max(sigma_biased), np.max(sigma_unbiased)), 100)
plt.figure(figsize=(15, 10))
plt.hist(sigma_biased, bins=bins, label = 'sigma_biased')
plt.hist(sigma_unbiased, bins=bins, label = 'sigma_unbiased')
plt.legend(loc = 'best')
plt.axvline(x=1, color = 'black')
plt.show()
```

```python
# Хм, а давайте посмотрим, как меняется среднее этой оценки в зависимости от кол-ва наблюдений
sigma_biased_mean = []
sigma_unbiased_mean = []
for j in range(10, 1000, 10):
    sigma_biased = []
    sigma_unbiased = []
    for i in range(1000):
        X = np.random.normal(loc = 0, scale = 1, size = j)
        sigma_biased.append(1/len(X)*np.sum((X-np.mean(X))**2))
        sigma_unbiased.append(1/(len(X)-1)*np.sum((X-np.mean(X))**2))
    sigma_biased_mean.append(np.mean(sigma_biased))
    sigma_unbiased_mean.append(np.mean(sigma_unbiased))
```

```python
plt.figure(figsize=(10,10))
plt.plot(sigma_biased_mean, label = 'biased')
plt.plot(sigma_unbiased_mean, label = 'unbiased')
plt.legend()
plt.show()
```

- Другими словами, состоятельной считается, оценка, которая при увеличении наблюдений, стремится к своему настоящему значению

Например, для формального определения состоятельности надо бы помнить про сходимость по вероятности, а для формального определения эффективности круто было бы знать про информационную матрицу Фишера (на самом деле это только звучит страшно)

### Предпосылки теоремы Гаусса-Маркова

$$Y = \alpha + \beta \cdot X + \varepsilon$$

1. Модель линейна по параметрам и правильно специфицирована
    - Это означает, что каждый член правой части включает коэффициенты (в данном случае $\alpha \text{ и } \beta$ ) как простой множитель, и здесь нет встроенных зависимостей между коэффициентами
    - Например, такая модель не является линейной по параметрам: $Y = \alpha \cdot X^{\beta} + \varepsilon$
2. $X$ - детерминированная величина (не путать с константой)
3. Математическое ожидание случайного члена равно 0: $E(\varepsilon) = 0$
4. Случайный член гомоскедастичен: $var(\varepsilon) = \sigma^2$, где $\sigma^2$ - некая константа
5. Значения случайного члена имеют взаимно независимые распределения: $E(\varepsilon_i \cdot \varepsilon_j)=0$ для всех $i \ne j$

- Коротко:
    $$E[\varepsilon] = 0, E[\varepsilon^2] = \sigma^2, E(\varepsilon_i, \varepsilon_j) = 0, i, j = 1, ..., N, i \ne j \\ 
    X - \text{детерменированы}$$

- Если все вышеперечисленные предпосылки выполняются, то оценка коэффициентов, полученная с помощью МНК, явлется наилучшей линейной несмещенной оценкой (НЛНО или BLUE - best linear unbiased estimator)
- Стоит заметить, что мы с вами ничего не сказали про распределение случайного члена
- Другими словами, для того, чтобы оценки, полученные с помощью МНК, были НЛНО, нам не надо предполагать какое-то конкретное распределение (например, нормальное) у случайного члена

- Конечно же, по какой-то причине, предполагается нормальность случайной компоненты
- Тогда, мы с вами еще упростить запись предпосылок: $\varepsilon_i \overset{i.i.d.}{\sim} N(0, \sigma^2)$, где `i.i.d.` - idependent identically distributed, т.е. независимые и одинаково распределенные
- Нам необходима предпосылка о нормальности для выполнения статистических тестов для гипотез и построения доверительных интервалов на основе результатов оценивания регресии.
